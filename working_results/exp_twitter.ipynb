```python
#save dataset to dic
a_dic={}
with open('twitter_combined.txt', 'rt') as f:
  for line in f:
    key,value = line.split()
    if key not in a_dic:
      a_dic[key] = value
    else:
      a_dic[key] = a_dic[key]+' '+value
    if value not in a_dic:
      a_dic[value] = key
    else:
      a_dic[value] = a_dic[value]+' '+key

f.close()
```
